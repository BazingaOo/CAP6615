```python
import pandas as pd
import numpy as np
```

```python
price=pd.read_csv('C:/Users/yiming.xu/Documents/GitHub/CAP6615/HW3/HistoricalPrices.csv')
pe=pd.read_csv('C:/Users/yiming.xu/Documents/GitHub/CAP6615/HW3/ie_data_PE.csv')
```

```python
price
```

```python
pe=pe.rename({'Date':'Date_PE'},axis=1)
pe
```

```python
price['Date_PE']=0.0
for i in range(price.shape[0]):
    month=int(price['Date'][i][0:2])
    year=int(price['Date'][i][-2:])
    if year>50:
        year=year+1900
    else:
        year=year+2000
    price['Date_PE'][i]=year+month*0.01
```

```python
data=pd.merge(price,pe,how='left',on='Date_PE')
data.to_csv('C:/Users/yiming.xu/Documents/GitHub/CAP6615/HW3/Price_PE.csv')
```
